```python
# Import libraries
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create 10 sentences (3 topics)
sentences = [
    # Cricket
    "The batsman hit a powerful six over long-on.",
    "The bowler delivered a fast yorker that knocked over the stumps.",
    "The cricket match went into a thrilling super over.",
    "The team celebrated after winning the cricket tournament.",

    # Cooking
    "She added garlic and olive oil to the pan before cooking the pasta.",
    "The chef baked a chocolate cake in the oven.",
    "Fresh vegetables were chopped to prepare a healthy salad.",

    # Cybersecurity
    "The company installed a firewall to protect its network.",
    "Cybersecurity experts detected a phishing attack in the email system.",
    "Strong passwords and two factor authentication improve online security."
]

print("Total sentences:", len(sentences))

# Step 2: Generate embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
print("Embedding shape:", embeddings.shape)

# Step 3: Compute cosine similarity matrix
similarity_matrix = cosine_similarity(embeddings)
print("Cosine Similarity Matrix:\n", similarity_matrix)

# Step 4: Display 10×10 heatmap
plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=True, cmap="coolwarm",
            xticklabels=range(1,11), yticklabels=range(1,11))
plt.title("Sentence Similarity Heatmap")
plt.show()

# Step 5: Query similarity search
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])
similarities = cosine_similarity(query_embedding, embeddings)[0]

top_indices = similarities.argsort()[-2:][::-1]

print("Query:", query)
print("\nTop 2 most similar sentences:\n")
for idx in top_indices:
    print(sentences[idx])
    print("Similarity Score:", similarities[idx])
    print()x    xticklabels
```
